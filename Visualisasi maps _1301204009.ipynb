# Import dataset covid indonesia

```python
covid_indo = pd.read_csv("/content/drive/MyDrive/semester 7/fiqri visdat/WEEK 10 /covid_19_indonesia_time_series_all.csv")
covid_indo
```

```python
# Convert the 'Location' column to uppercase
covid_indo['Location'] = covid_indo['Location'].str.upper()

# Show the first few rows of the dataframe to verify the changes
covid_indo[['Location']]
```

```python
cv1jun = covid_indo["Date"] == "6/1/2021"
cv1jun
```

```python
cvindo = indo.merge(covid_indo[cv1jun], left_on="Propinsi", right_on="Location")
cvindo
```

```python
ax = gplt.choropleth(
    cvindo,
    hue="New Cases",
    edgecolor="white",
    linewidth=5,
    cmap="Reds",
    alpha = 1,
    projection=gcrs.AlbersEqualArea(),
    figsize=(30, 30)
)

gplt.polyplot(
    indo,
    ax=ax,
    extent=indo.total_bounds,
    edgecolor="black",
    linewidth=1,
    zorder=1
)

ax.set_title(
    "Sebaran Kasus Positif COVID19 di Indonesia per Tanggal 1 Juni 2021",
    fontsize=36
)
```
